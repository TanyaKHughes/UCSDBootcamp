## **Web scraping data using pandas**

The following block of code will retrieve the "2018-19 NBA Player Stats: Per Game" data from http://www.basketball-reference.com/.

```python
import pandas as pd

#Get data
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
html = pd.read_html(url, header = 0)
df = html[0]   # the 2019 data
df
```

## **Data Cleaning**

Some of the age values are set to "Age" instead of an actual number, as you can see:

```python
n = 0
for age in df.Age:
  print(age)
  n+=1
  if n>30: break

```

So we need to remove those rows with bad age values:

```python
raw = df.drop(df[df.Age == "Age"].index)
n = 0
for age in raw.Age:
  print(age)
  n+=1
  if n>30: break
```

```python
raw
```

```python
raw.shape
```

```python
raw.head()
```

```python
# Find how many null values we have
raw.isnull().sum()  # isnull() is the same as isna() which returns a same-sized array with True or False depending on whether it has NaN, NaT, or None.
                    # a blank in the original table will be False according to these methods however.
```

```python
# replace NA's with 0's
df = raw.fillna(0)
df.isnull().sum()
```

```python
# Get rid of the redundant Rk column
df.drop(['Rk'], axis=1, inplace=True)
df
```

```python
df.to_csv("NBAdata.csv", index=True)  # Write to a file, including the row indices (this is the default for index)
! ls
```

# **QUESTIONS**

### Which player scored the most points per game?  How many did he score?

```python
highest_scoring_player = df[df.PTS == df.PTS.max()]
indx = highest_scoring_player.index[0]
print(f"{highest_scoring_player.at[indx,'Player']} scored {highest_scoring_player.at[indx,'PTS']} points.")
```

```python
highest_scoring_player
```

What team is he from?  What position does he play?  How many games did he play in?

```python
print(f"{highest_scoring_player.at[indx,'Player']} is from the {highest_scoring_player.at[indx,'Tm']} team" \
          f" and plays position {highest_scoring_player.at[indx,'Pos']}." \
      f" He played in {highest_scoring_player.at[indx,'G']} games.")
```

### Which players scored more than 20 points per game?

```python
high_scoring_df= df[df.PTS>20]
high_scoring_df
```

```python
print("The players who scored over 20 pts/game are")
for player in high_scoring_df['Player'].values:
  print(player)

```

### Which player(s) had the most 3-ptrs per game?

```python
high_3P_scorers = df[df['3P'] == df['3P'].max()]
print(f"The player(s) with the highest # 3-ptrs per game is(are)")
for player in high_3P_scorers["Player"].values:
  print(player)
```

### Which player(s) had the most assists per game?

```python
high_assists_df = df[df['AST'] == df['AST'].max()]
print(f"The player(s) with the highest # assists per game is(are)")
for player in high_assists_df["Player"].values:
  print(player)
```
